```python
import pandas as pd
import numpy as np
import re  

#Import useful libraries that will aid in data processing
```

```python
df_2018 = pd.read_csv('startup_funding2018.csv')   #To get the data into python to be processed and analysed
df_2018
```

```python
df_2018['Amount'] = df_2018['Amount'].replace(",","",regex = True)  # Clear the commas in the Amount column to make sure your figures are uniform
```

```python
df_2018
```

```python
df_2018['Amount'] = df_2018['Amount'].replace("—",0)

df_2018

# Replace - with zero in the amount column
```

```python
df_2018['INR_Amount'] = df_2018['Amount'].str.rsplit('₹', n=2).str[1]

df_2018

# Create a new column for the Rupee figures so it can be converted to USD
```

```python
df_2018['INR_Amount'] = df_2018['INR_Amount'].apply(float).fillna(0)

df_2018['USD_Amount'] = df_2018['INR_Amount']*0.012

#Make sure the new Rupee column is a float and use the rate of 0.012 to convert it to USD into a new column
```

```python
df_2018
```

```python
df_2018['USD_Amount'] = df_2018['USD_Amount'].replace(0,np.nan)

df_2018['USD_Amount'] = df_2018['USD_Amount'].fillna(df_2018['Amount'])

df_2018

#Replace the 0 columns in the USD to na 
#Fill the na with the rest of the figures in the original amount column which are already in dollars
```

```python
df_2018['USD_Amount'].value_counts()

#Checking for any disparity that migh require further attention in the USD Amount column and it was noticed some figures had $ attached to it
```

```python
df_2018['USD_Amount'] = df_2018['USD_Amount'].replace("$","",regex = True)

df_2018

#Replace the $ with nothing and display the dataframe
```
