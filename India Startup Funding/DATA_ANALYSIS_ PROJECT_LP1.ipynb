## HYPOTHESIS TESTING

```python
# OUR SIGNIFICANCE LEVEL IS 5%

# WE ARE USING T TEST

#perform two sample t-test

yr = cleaned_new_data['Year of Funding']
amt = cleaned_new_data['Amount($)']

t,p = stats.ttest_ind(a=yr, b=amt) 

print(f't-value: {t}')
print(f'p-value: {p}')
```

```python
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
#sns.histplot(data=cleaned_new_data.iloc[:10], x='Amount($)', hue='Year of Funding', kde=True)
#plt.show()
```

### Since the p-value of the test (0.0541) is equal to or greater than .05, we accept the null hypothesis.

### This means we have sufficient evidence to say that the Year of Funding is not dependant on the Amount raised.
